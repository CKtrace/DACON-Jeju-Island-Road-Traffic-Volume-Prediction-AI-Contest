```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df1 = pd.read_csv("test.csv")
df.head()
```

```python
df = df.drop('id', axis=1)
df1= df1.drop('id', axis=1)
df1
```

## 20221017 -> 2022-10-17 변환 작업

```python
df['datetime'] = df['base_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
df1['datetime'] = df1['base_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
df.head()
```

```python
df1.head()
```

```python
df.drop('base_date', axis=1, inplace=True)
df1.drop('base_date', axis=1, inplace=True)
df1
```

## 연 월 일 분리

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day

df1['year'] = df1['datetime'].dt.year
df1['month'] = df1['datetime'].dt.month
df1['day'] = df1['datetime'].dt.day

df.head()
```

```python
df.drop('datetime', axis=1, inplace=True)
df1.drop('datetime', axis=1, inplace=True)
df.head()
```

## 제한 구역(있음/없음) -> One-Hot Encoding 

```python
df = pd.get_dummies(df, columns=['start_turn_restricted', 'end_turn_restricted'])
df1 = pd.get_dummies(df1, columns=['start_turn_restricted', 'end_turn_restricted'])
```

```python
df1
```

```python
df.columns
```

## day_of_week, road_name, start_node_name, end_node_name 변환

```python
from sklearn.preprocessing import LabelEncoder

class_le = LabelEncoder()
y = class_le.fit_transform(df['day_of_week'].values)
x = class_le.transform(df1['day_of_week'].values)
df['day_of_week']=y
df1['day_of_week']=x
df.head()
```

```python
class_le1 = LabelEncoder()

df['road_name'] = class_le1.fit_transform(df['road_name'].values)
df1['road_name'] = class_le1.transform(df1['road_name'].values)

class_le2 = LabelEncoder()

df['start_node_name'] = class_le2.fit_transform(df['start_node_name'].values)
df1['start_node_name'] = class_le2.transform(df1['start_node_name'].values)

class_le3 = LabelEncoder()

df['end_node_name'] = class_le3.fit_transform(df['end_node_name'].values)
df1['end_node_name'] = class_le3.transform(df1['end_node_name'].values)

df
```

```python
df.to_csv("preprocess_train.csv")
df1.to_csv('preprocess_test.csv')
```
