```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("preprocess_train.csv", index_col=0)
df.head()
```

```python
sns.scatterplot(x=df['start_latitude'], y=df['end_latitude']) 
```

```python
sns.scatterplot(x=df['start_longitude'], y=df['end_longitude']) 
```

```python
sns.scatterplot(y=df['weight_restricted'], x=df['road_type']) 
```

```python
df.drop('end_latitude', axis=1, inplace=True)
df.drop('end_longitude', axis=1, inplace=True)
df
```

```python
df.rename(columns={'start_turn_restricted_없음':'start_turn_restricted_X', 'start_turn_restricted_있음':'start_turn_restricted_O',
         'end_turn_restricted_없음':'end_turn_restricted_X', 'end_turn_restricted_있음':'end_turn_restricted_O'}, inplace=True)
df.head()
```

```python
df.to_csv('11_01_train.csv')
```
